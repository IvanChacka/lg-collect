# hot-collect 工作流可视化（Jupyter）

这个 Notebook 用于在开发时快速生成/查看 LangGraph 的 Mermaid 工作流图。

前置：确保你在项目根目录运行，并已安装 Jupyter（例如 `pip install notebook`）。


```python
import sys
from pathlib import Path

# Notebook 默认工作目录是 notebooks/，这里把项目根目录加到 sys.path，保证能 import core/nodes/tools
PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from core.viz import mermaid

m = mermaid()
print(m)
```

```python
import json
from IPython.display import HTML

# 用前端脚本把 Mermaid 源码塞进 div，再渲染（避免 HTML 转义问题）
payload = json.dumps(m, ensure_ascii=False)
HTML(
    f"""
    <div id='graph' class='mermaid'></div>
    <script type='module'>
      import mermaid from 'https://cdn.jsdelivr.net/npm/mermaid@10/dist/mermaid.esm.min.mjs';
      mermaid.initialize({{ startOnLoad: false }});
      const src = {payload};
      const el = document.getElementById('graph');
      el.textContent = src;
      await mermaid.run({{ querySelector: '.mermaid' }});
    </script>
    """
)
```
